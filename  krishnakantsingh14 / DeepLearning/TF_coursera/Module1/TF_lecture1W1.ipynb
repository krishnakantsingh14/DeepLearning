Let's assume we have x, y apir of data. Our goal is to find a funnction that maps x to y, i.e.
y = f(x)

X = -1, 0, 1, 2, 3, 4

y = -3, -1, 1, 3, 5, 7




```
import tensorflow as tf 
import tensorflow.keras as keras
import tensorflow.keras.layers as layers
import numpy as np 
import pandas as pd
import pprint 
```

```
def getData():
    """ Generate and return data for the neural network"""
    xs = np.array([-1, 0, 1, 2, 3, 4], dtype=float)
    ys = np.array([-3, -1, 1, 3, 5, 7], dtype =float)
    return xs, ys 
```

```
def defineModel():
    model = keras.Sequential([
                            layers.Dense(units=1, activation=None, input_shape=[1])
    ])
    return model

def compileModel(model, optimizer, loss):
    model.compile(optimizer=optimizer, loss=loss)
```

```
xs, ys = getData()
```

```
model = defineModel()
compileModel(model, optimizer='sgd', loss="mean_squared_error")
```

## Training the neural network

```
history = model.fit(xs,ys,epochs=500)
```

```
pd.DataFrame(history.history).plot()
```

```
for layer in model.layers:
    print ('W: {}'.format(layer.weights[0].numpy()))
    print ('b: {}'.format(layer.weights[1].numpy()))

```

```
model.predict(np.array([10]))
```
